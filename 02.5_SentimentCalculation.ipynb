# Sentiment Analysis of Tech Company Tweets

```python
# importing libraries
import os
import pandas as pd
```

```python
# reading in data
Tweet = pd.read_csv("Tweet.csv")
Tweet.head()
```

```python
# importing the VADER sentiment analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# filtering out tweets with engagement less than 5 
Tweet2 = Tweet[Tweet['engagement']>=5]
# cleaning tweet body 
Tweet2['prep_body'] = Tweet2['body'].replace(r"https?:\S+|http?:\S+|www?:\S+", '', regex=True).replace(r"[@#\$][a-zA-Z]+", '', regex=True).replace(r"\s\s+", ' ', regex=True).str.strip()
```

```python
# calculating sentiment scores

compound=[]
positive=[]
negative=[]

for i in range(0, len(Tweet2)):
    compound.append(analyzer.polarity_scores(Tweet2.iloc[i]['prep_body'])['compound'])
    positive.append(analyzer.polarity_scores(Tweet2.iloc[i]['prep_body'])['pos'])
    negative.append(analyzer.polarity_scores(Tweet2.iloc[i]['prep_body'])['neg'])
```

```python
Tweet2['compound']=compound
Tweet2['positive']=positive
Tweet2['negative']=negative
Tweet2.head()
```

```python
Tweet2.to_csv('SentimentTweet2.csv')
```
